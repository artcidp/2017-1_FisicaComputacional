# Problema 2

```python
def fun(x):
    return x**3-5*x+2

def der_analitica(x):
    return 3*x**2-5

def der_numerica(funcion,x,h=1e-3):
    """
    Realiza la derivada de la funcion por diferencias finitas dado el punto a evaluar y la diferencia h 
    """
    df=(funcion(x+h)-funcion(x))/h
    return df
```

```python
x=np.linspace(-5,5,200)
fig=plt.figure(figsize=(10,10))

plt.plot(x,fun(x),label="funcion")
plt.plot(x,der_analitica(x),label="der. analitica",lw=2)
plt.plot(x,der_numerica(fun,x,h=.1),label="der. analitica",alpha=0.5,lw=3) #ajustamos la transparencia con alpha

plt.legend()
plt.grid()
plt.show()
```

Observamos que ambas derivadas son semejantes :)

# Problema 3

```python
def Metodo_de_Newton(fun,der,p0,nmax=100):
    x=p0
    for i in xrange(nmax):
        x=x-fun(x)/der(x)
    return x
```

```python
def foo(x):
    return np.sin(3*x)+0.5*np.cos(2*x)**2

def der_foo(x):
    return 3*np.cos(3*x)-2*np.cos(2*x)*np.sin(2*x)
```

```python
x=np.linspace(-6,6,200) #nos limitaremos al intervalo -6,6
raices=Metodo_de_Newton(foo,der_foo,x,1000) #sacamos raices con el medoto de Newton variando las condiciones inicales en x

fig=plt.figure(figsize=(10,10))

plt.plot(x,foo(x),label="funcion")
plt.plot(x,der_foo(x),label="der. analitica",ls='dotted',lw=3)
plt.plot(raices,np.zeros(len(raices)),label="raices",marker='o')

plt.xlim([-6,6])
plt.legend()
plt.grid()
plt.show()
```

# Problema 4

```python
def bessel_0(x):
    return jv(0,x)

def der_analitica_bessel(x):
    df=-jv(1,x)
    return df

def der_numerica_bessel(x,h=.1):
    df=der_numerica(bessel_0,x,h)
    return df
```

```python
x=np.linspace(0,10,200) #nos limitaremos al intervalo 0,10
raices_analiticas=Metodo_de_Newton(bessel_0,der_analitica_bessel,x,100)
raices_numericas=Metodo_de_Newton(bessel_0,der_numerica_bessel,x,100)

fig=plt.figure(figsize=(10,10))

plt.plot(x,jv(0,x),label="funcion",color="black")
plt.plot(x,-jv(1,x),label="der. analitica",ls='dotted',lw=3,color="red")
plt.plot(raices_analiticas,np.zeros(len(raices_analiticas)),label="raices analiticas",marker='x',color="red",markersize=15)

plt.plot(x,-jv(1,x),label="der. analitica",ls='-.',lw=3,color="blue")
plt.plot(raices_numericas,np.zeros(len(raices_numericas)),label="raices numericas",marker='o',color="blue")

plt.xlim([0,10])
plt.legend()
plt.grid()
plt.show()
```

# Problema 5

El inverso de una matriz de 2x2 esta dado por:
$$
\left(\begin{array}{cc}
a&b\\
c&d
\end{array}\right)^{-1}
=
{1 \over a d - b c}
\left(\begin{array}{rr}
d&-b\\
-c&a
\end{array}\right)~,\text{si }\ ad-bc \ne 0.
$$


```python
def inversa_2x2(matriz):
    """
    Toma la matriz como una lista de la forma [[a,b],[c,d]] y regresa su inversa si el determinante es no cero
    """
    a=matriz[0][0]
    b=matriz[0][1]
    c=matriz[1][0]
    d=matriz[1][1]
    det=a*d-b*c
    if det!=0:
       return [[d/det,-b/det],[-c/det,a/det]]
    else:
        print "No tiene matriz inversa"
```

```python
inversa_2x2([[1,0],[0,0]]) #ejemplo de determinante cero
```

```python
inversa_2x2([[4.,0],[0,2.]]) #ejemplo de inversa de matriz diagonal
```

# Problema 6

```python
def suma_digitos(x): 
    string=str(x)
    suma=0 #inicializamos la suma
    for i in xrange(len(string)):
        if string[i]!='.' and string[i]!='-':
            suma+=int(string[i])
    return suma
```

```python
suma_digitos(-1.502) #si el numero es negativo, no consideramos el signo
```

```python
suma_digitos(1e-3) #con notacion cientifica
```

```python
print suma_digitos(np.pi)
print np.pi
```
